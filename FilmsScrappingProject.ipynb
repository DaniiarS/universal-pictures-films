## Gets movies by the year of release

```python
import requests
from bs4 import BeautifulSoup

def add_to_movie_list( movie_list, movie ):
    movie_list.append( movie )
    return movie_list

def add_movie_info( movie, title ):
    movie['title'] = title


movie_list_page = requests.get('https://en.wikipedia.org/wiki/Lists_of_Universal_Pictures_films')
soup = BeautifulSoup( movie_list_page.content, 'html.parser' )

base_link = 'https://en.wikipedia.org'

movie_list = []
for link in soup.find( 'div', class_ = 'mw-body-content mw-content-ltr' ).find('ul').find_all('li'):
    # Go through links 
    # Each link contains a list of all films for the given year
    
    # Generate a link
    movie_list_link = base_link + movie_list_link.find('a')['href']
    
    page = requests.get( movie_list_link )
    soup2 = BeautifulSoup( page.content , 'html.parser' )

    for table in soup2.find_all( 'table' ):
        for td in table.find( 'tbody' ).find_all( 'i' ):
            movie_info = {}
            try:
                title = td.find( 'a' ).get_text()
                add_movie_info( movie_info, title )
                add_to_movie_list( movie_list, movie_info )
            except Exception as e:
                pass
```

## Connects to the OMDB API and makes Requests

```python
import json

def save_data( title, data ):
    with open( title, 'w', encoding = 'utf-8' ) as f:
        json.dump( data, f, ensure_ascii = False, indent = 2 )
```

```python
import json

def load_data( title ):
    with open( title, 'r' ) as f:
        data = json.load(f)
    return data
```

```python
import requests
import urllib
import os

def get_movie_omdb( title ):
    base_url = 'http://www.omdbapi.com/?'
    parameters = { 'apikey': '9d7bd21', 't': title }
    encoded_url = urllib.parse.urlencode( parameters )
    
    url = base_url + encoded_url
    return requests.get( url ).json()
```

```python
movie_list = load_data( 'universal_pictures_films.json' )

count = 0
for movie in movie_list:
    if count == 10:
        break
    if len(movie.keys()) == 1:
        print(movie)
        count += 1
```
